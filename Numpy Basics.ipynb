```python
from functools import reduce
array = {1, 2, 3}
result = lambda array: reduce(lambda P, x: P + [subset | {x} for subset in P], array , [set()])

print(result(array))

```

```python
import pandas as pd
df1 = pd.DataFrame({"name": ['tobey', 'peter', 'chris', 'pratt'],'chem_score':[10, 9, 8, 7], 'phy_score':[7, 8, 9, 10], 'Hindi_score':[9, 9, 9, 9]})
df2 = pd.DataFrame({"name": ['chris', 'pratt', 'andrew', 'tom'],'chem_score':[10, 10, 10, 9], 'maths_score':[6, 6, 7, 9], 'eng_score':[9, 10, 10, 9]})

df1, df2
```

```python
dfn=pd.concat([df1, df2], axis=0)
dfn
```

```python
dfn.isna().sum()
dfn.isna().sum().sum()
dfn.isna().any()
```

```python
dfn.fillna(0)
```

```python
# .join happens on indexes
dfj=df1.join(df2, how='inner', lsuffix='_left', rsuffix='_right')
dfj
```

```python
# .merge happens on columns as sql inner join
dfm=df1.merge(df2, how='inner', on='name')
dfm
```

```python
dfm.mean()
```

```python
dfm.drop('name', axis=1)
dfm.drop('name', axis=1).mean()   #this can be done only when all the columns data of df is numeric.
```
